# B) Uni-Variate EDA

Statistical or visual analysis of a single column (one variable)

### 1. Find value counts of categorial columns namely Category, Segment, Sub Category, Region,Ship Model and Market .
###  Depict the following

a) Category count on a bar chart(matplotlib) <br>
b) Sub-Category on a horizontal bar chart (matplotlib)<br>
c) Segment on a Pie Chart(matplotlib)<br>
d) Region on a bar chart(seaborn)<br>
e) Ship Mode on a line chart(matplotlib)<br>
f) market on a area chart(matplotlib)

```python
cat_cols
```

```python
a1 = df["Category"].value_counts()
a1
```

```python
plt.bar(a1.index,a1.values,color="lightgreen",edgecolor="black",label="Category counts")
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Category Count - Vertical Bar Chart")
plt.legend()
plt.show()
```

```python
a2 = df["Sub-Category"].value_counts()
a2
```

```python
plt.barh(a2.index,a2.values,color="blue",label="Sub Category counts")
plt.xlabel("Sub-Category")
plt.ylabel("Count")
plt.title("Sub-Category Count - Vertical Bar Chart")
plt.legend()
plt.show()
```

```python
a3 = df["Segment"].value_counts()
a3
```

```python
plt.pie(a3.values,labels=a3.index,autopct="%.2f%%")
plt.title("Segment Count Percentage distirbution")
plt.legend(loc=1)
plt.show()
```

```python
a4 = df["Region"].value_counts()
print(type(a4)) #series
a4
```

```python
plt.bar(a4.index,a4.values,color="orange",edgecolor="black",label="Region counts")
plt.xlabel("Region")
plt.ylabel("Count")
plt.title("Region Count")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x=df["Region"])
plt.title("Region Count - Vertical Bar Chart")
plt.xticks(rotation=90)
plt.show()
```

```python
a5 = df["Ship Mode"].value_counts()
a5
```

```python
plt.plot(a5.index,a5.values,color="blue",marker="o",label="Ship Mode counts")
plt.xlabel("Ship Mode")
plt.ylabel("Count")
plt.title("Ship Mode Count")
plt.legend()
plt.show()
```

```python
a6 = df["Market"].value_counts()
a6
```

```python
plt.stackplot(a6.index,a6.values)
plt.xlabel("Market")
plt.ylabel("Count")
plt.title("Market Count - Area Chart")
plt.show()
```

## 2) Plot the following

## c) Bivariate Analysis
Statistical or Visual Analysis of 2 variables

## 1) Depict the following on a Scater plot
a) Sales vs Profit <br>
b) Profit vs Shopping Cost

```python
plt.scatter(df["Sales"],df["Profit"],color = "green")
plt.title("Sales vs Profit - Scatter Chart")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
plt.scatter(df["Profit"],df["Shipping Cost"],color = "tan",marker="*")
plt.title("Profit vs Shipping Cost - Scatter Chart")
plt.xlabel("Profit")
plt.ylabel("Shipping Cost")
plt.show()
```

## 2) Depict Boxplot for the following
### a) Quantity based on Segment
### b) Discount based on Ship Mode

```python
sns.boxplot(x=df["Quantity"],y=df["Segment"])
plt.title("Boxplot - Quantity based on Segment")
plt.show()
```

```python
sns.boxplot(x=df["Discount"],y=df["Ship Mode"])
plt.title("Boxplot - Quantity based on Ship Mode")
plt.show()
```

## Depict the following

### a) Discount vs Market on a violinplot
### b) Quantity vs Category on a violinplot

```python
sns.violinplot(x=df["Discount"],y=df["Market"])
plt.title("Violinplot - Discount based on Market")
plt.show()
```

```python
sns.violinplot(x=df["Quantity"],y=df["Category"])
plt.title("Violinplot -Quantity based on Category")
plt.show()
```

### 4) Plot pairplot for the categorial variables including ["Sales","Profit","Quantity","Discount"]
